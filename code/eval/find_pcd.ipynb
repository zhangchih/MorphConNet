```python
import os
import shutil
import string

glia_file = '/code/PyTorch-BYOL/eval/neuron/neuron.txt'
t_file = '/code/PyTorch-BYOL/eval/neuron/neuron_.txt'
path = '/data/nerdxie/neuron_cluster/neuron_dataset_add_all_128/cluster3/'
glia_path = '/code/PyTorch-BYOL/eval/neuron/'

f = open(glia_file, "r")  # 设置文件对象
t = open(t_file, "w")

line = f.readline()
line = line[:-1]
i = 1
while line:  # 直到读取完文件
    print(i)
    if i is not 1:
        line = f.readline()  # 读取一行文件，包括换行符
    if line.find('_')>=0:
        line = line.split('_')[1]
    id = str("".join(list(filter(str.isdigit, line))))
    file_path = path + 'cluster3_' + id +'.txt'
    if os.path.exists(file_path):
        target = 'neuron_' + id +'.txt'
        shutil.copy(file_path, os.path.join(glia_path, target))
        t.write(target+'\n')
        print(id)
    else:
        print(file_path)
    i = i+1
f.close()

# source_path = '/output/glia'
# target_path = '/output/glias/'
# s = os.listdir(source_path)
# for file in s:
#     print(file)
#     shutil.copy(os.path.join(source_path, file), os.path.join(target_path, file.replace('cluster3', 'glia')))
```

```python
import os
import shutil
import string

# 替换文件字符
s_file = '/code/PyTorch-BYOL/eval/data/axon_terminal.txt'
t_file = '/data/chih0321/FAFB_Subcompartment_correct/axon_terminal.txt'
s = open(s_file, "r", encoding='utf-8')
line = s.readline()
fp = open(t_file, 'w', encoding='utf-8')
while line:
    fp.write(line.replace(', ', '\n'))
    line = s.readline()
fp.close()
```
